```python
import matplotlib.pyplot as plt
import numpy as np
all_walks =[]

for x in range(3400):
    random_walk = [0]
    for i in range(10):
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        
        if dice <= 3 :
            step = max(0, step - dice)
        elif dice <= 5 :
            step = step + 3
        else :
            step = step + dice
        if np.random.rand() <= 0.001 :
            step = 0
      
        
        random_walk.append(step)            
    all_walks.append(random_walk)
    
print(all_walks)
np_aw = np.array(all_walks)
np_aw_t = np.transpose(np_aw)
plt.plot(np_aw_t)
plt.show()
np.amax(all_walks)
```
